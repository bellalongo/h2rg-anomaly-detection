```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
goal_h = 2048
goal_w = 2048
```

Figure out what is wrong with 20240409_163902_euc_15db_noise

```python
root_dir = '/proj/case/2025-06-05/FPM_EM3_EM6_pre_vibe/20240409_163902_euc_15db_noise'

filenames = os.listdir(root_dir)
```

```python
img = Image.open(f'{root_dir}/{filenames[0]}')
plt.figure(figsize=(10, 8))
plt.imshow(img, cmap = 'Greys')
plt.axis('off')  # Hide axes
plt.title(filenames[0])
plt.show()

# Make the image into an array and grab the height and the width
img_array = np.array(img)
h, w = img_array.shape

print(img_array.shape)

# Calculat the midpoint and split the array in half
mid_point = w // 2
d1 = img_array[:, :mid_point]
d2 = img_array[:, mid_point:]

# Calculate how many cols to cut off
cols_to_cut = (mid_point - goal_w) // 2

# Remove 1 from the right as well

d1 = d1[:, cols_to_cut:-cols_to_cut]
d2 = d2[:, cols_to_cut:-cols_to_cut]

print(d1.shape, d2.shape)

# Plot both detectors
fig, axs = plt.subplots(2, 1, figsize = (10,12))

# Detector 1
im1 = axs[0].imshow(d1, cmap = 'Greys', origin = 'lower')
axs[0].set_title('Full Image Detector 1 (Left)')
axs[0].set_xlabel('X (Pixels)')
axs[0].set_ylabel('Y (Pixels)')
plt.colorbar(im1, ax = axs[0], label = 'Counts')

# Detector 3
im1 = axs[1].imshow(d2, cmap = 'Greys', origin = 'lower')
axs[1].set_title('Full Image Detector 2 (Right)')
axs[1].set_xlabel('X (Pixels)')
axs[1].set_ylabel('Y (Pixels)')
plt.colorbar(im1, ax = axs[1], label = 'Counts')

plt.show()

img.size
```
